```python
# was made with Amie for frictionless version and damped version was added by me
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

def findNewPos(vel, pos, dt):
    return(vel*dt + pos)

def findNewVel(accel, vel, dt):
    return(accel*dt + vel)

def findNewAccel(P, pos, K, vel):
    return(-(P*pos) - (K*vel))

def plotSpring(time, vel, pos, P, K, num):
    positions = []
    currentVel = vel;
    currentPos = pos
    time = [t/10 for t in range(1,1000)]
    dt = 0.1
    
    for t in time:
        currentAccel = findNewAccel(P, currentPos, K, currentVel)
        currentVel = findNewVel(currentAccel, currentVel, dt)
        #print(currentVel)
        currentPos = findNewPos(currentVel, currentPos, dt)
        #print(currentPos)
        positions.append(currentPos)
    
    #creating subplots with data
    plt.subplot(1,2,num+1)
    plt.plot(time, positions)
    plt.xlabel('Time (s)')
    plt.ylabel('Positon (m)')
    
    # setting the title for the subplots
    if (num == 0):
        plt.title('Frictionless')
    else:
        plt.title('Damped oscillator')
    plt.plot(time, positions)
    plt.tight_layout()
    return positions


def main():
    time = [t/10 for t in range(1,1000)]
    
    # data for frictionless and damped oscillator version
    # plotSpring accepts variables in order (time, initial velocity, initial position, P constant,
    #                                        K constant, and number of subplot)
    frictionlessData = plotSpring (time, 0, 5, 0.1, 0, 0)
    dampedData = plotSpring(time, 0, 5, 0.1, 0.1, 1)
    
    # plotting the combined plot
    plt.figure(3)
    plt.plot(time, frictionlessData, label = 'Frictionless')
    plt.plot(time, dampedData, label = 'Damped oscillator')
    plt.legend()
   
if __name__ == "__main__":
    main()
```
