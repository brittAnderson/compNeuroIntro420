```python
v_0 = 0
s_0 = .5
p = 10
k = 1

times = []
locations = []

t = 1

vel = v_0
loc = 2
diff_t = 0.1

while (t < 10):
    acc = -p * loc
    vel = vel + acc
    dist = vel * diff_t
    loc = loc + dist 
    times.append(t)
    locations.append(loc)
    t += 0.1

##########

times_damped = []
locations_damped = []

t = 1

while (t < 10):
    acc = -p * loc - k * vel * diff_t
    vel = vel + acc
    dist = vel * diff_t
    loc = loc + dist 
    times_damped.append(t)
    locations_damped.append(loc)
    t += 0.1

%matplotlib inline
import matplotlib.pyplot as plt
plt.subplot(1, 2, 1)
plt.plot(times, locations)
plt.title('A Frictionless Spring')
plt.xlabel('time')
plt.ylabel('location')

plt.subplot(1, 2, 2)
plt.plot(times_damped, locations_damped)
plt.title('A Damped Oscillator Spring')
plt.xlabel('time')
plt.ylabel('location')
plt.tight_layout()
plt.show()
```
